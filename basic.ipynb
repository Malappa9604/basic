```python
numbers = [1,2,3,4]

square = []
for n in numbers:
    square.append(n*n)
print(square)

# map with this funcation
def square(x):
    return x*x
    numbers = [1,2,3,4]
    square = list(map(square, numbers))
    print(square)

    # map with this funcation
    # def square(x):
    #   return x*x
    # numbers = [1,2,3,4]
    # result = list(map(square,numbers))

    # map with lambda
    numbers = [1,2,3,4]
    square = list(map(lambda x: x*x, numbers))
    print(square)
```

```python
numbers = [1,2,3,4,5,6]

evens = []
for n in numbers:
    if n % 2 == 0:
        evens.append(n)
print(evens)

# method with filter
def is_even(x):
    return x % 2 == 0
numbers = [1,2,3,4,5,6]
result = list(filter(is_even, numbers))
print(result)


numbers = [1,2,3,4,5,6]
result = list(filter(lambda x: x % 2 == 0, numbers))
print(result)
```

```python
numbers = [1,2,3,4]

total = 0
for n in numbers:
    total += n
print(total)

# with reduce

from functools import reduce
def add(x,y):
    return x + y
numbers = [1,2,3,4]
result = reduce(add, numbers)


# with lambda
numbers = [1,2,3,4]
result = reduce(lambda x,y: x + y, numbers)
print(result)
```
